This calls the longest common sequences at the start of the sequence between two strings

```python
def longestCommonPrefix(s1, s2):
    i = 0
    while i < len(s1) and i < len(s2) and s1[i] == s2[i]: 
        i += 1
    return s1[:i]
longestCommonPrefix('ACCATGT', 'ACCAGAC')
```

```python
def match (s1, s2):
    if not len(s1) == len(s2):
        return False
    for i in range(len(s1)):
        if not s1[i] == s2[i]:
            return False
        
    return True
match('ATGCT', 'ATGCT')
```

This loop intreprets the quality score variables in a fasta file

```python
def readFastq(filename):
    sequences = []
    qualities = []
    with open(filename) as fh:
        while True: 
            fh.readline()
            seq = fh.readline().rstrip()
            fh.readline()
            qual = fh.readline().rstrip()
            if len(seq) == 0:
                break
            sequences.append(seq)
            qualities.append(qual)
    return sequences, qualities
seqs, quals = readFastq('SRR835775_1.first1000.fastq')
```

```python
def phred33ToQ(qual):
    return ord(qual) - 33
```

```python
phred33ToQ('J')
```

This creates a histogram of the frequencies of the different quality scores

```python
def createHist(qualities):
    hist = [0] * 50
    for qual in qualities:
        for phred in qual:
            q = phred33ToQ(phred)
            hist[q] += 1
    return hist
h = createHist(quals)
print(h)
```

Generates a histogram in matplot

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(len(h)), h)
plt.show()
```

Plot the C+G content - to see if the mix of different bases changes aross the read. IF there is a partiuclarly bad sequencing cycle there might be a high C+G content

```python
def findGCByPos(reads):
    gc = [0] * 100
    totals = [0] * 100
    
    for read in reads:
        for i in range(len(read)):
            if read[i] == 'C' or read[i] == 'G':
                gc[i] += 1
            totals[i] += 1
    for i in range(len(gc)):
        if totals[i] > 0:
            gc[i] /= float(totals[i])
    return gc

gc = findGCByPos(seqs)
plt.plot(range(len(gc)), gc)
plt.show()
```

Write the naive exact matching argument
eg. p = read sequence and t could be a genome that you want to match against

```python
def naive(p, t):
    occurrences = []
    for i in range(len(t) - len(p) + 1):
        match = True
        for j in range(len(p)):
            if not t[i+j] == p[j]:
                match = False
                break
        if match:
            occurrences.append(i)
    return occurrences
```

```python
import random
def generateReads(genome, numReads, readLen):
    ''' Generate reads from random positions in the given genome. '''
    
    reads = []
    for _ in range(numReads):
        start = random.randint(0, len(genome)-readLen) - 1
        reads.append(genome[start : start+readLen])
    return reads
```

```python
reads = generateReads(genome, 100, 100)

numMatched = 0
for r in reads:
    matches = naive(r, genome)
    if len(matches) > 0:
        numMatched += 1
print('%d / %d reads matches exactly' % (numMatched, len(reads)))
```

```python
def readFastq(filename):
    sequences = []
    qualities = []
    with open(filename) as fh:
        while True: 
            fh.readline()
            seq = fh.readline().rstrip()
            fh.readline()
            qual = fh.readline().rstrip()
            if len(seq) == 0:
                break
            sequences.append(seq)
            qualities.append(qual)
    return sequences, qualities
```

```python
phix_reads, _ = readFastq('ERR266411_1.first1000.fastq')
```

```python
numMatch = 0
n = 0
for r in phix_reads:
    r = r[:30]
    matches = naive(r, genome)
    n += 1
    if len(matches) > 0:
        numMatched += 1
print('%d / %d reads matched the genome!' % (numMatched, n))
```

```python
def reverseComplement(s): 
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A', 'N': 'N'}
    t = ''
    for base in s:
        t = complement[base] + t
    return t
```

```python
numMatch = 0
n = 0
for r in phix_reads:
    r = r[:30]
    matches = naive(r, genome)
    matches.extend(naive(reverseComplement(r), genome))
    n += 1
    if len(matches) > 0:
        numMatched += 1
print('%d / %d reads matched the genome!' % (numMatched, n))
```
